```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, AgglomerativeClustering
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
X, y = make_blobs(n_samples = 1000, centers=4, cluster_std=1.0, random_state=42)
print(X.shape)
print(y.shape)
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
wcss = []
for i in range(1,11):
  km = KMeans(n_clusters=i)
  km.fit_predict(X)
  wcss.append(km.inertia_)
  print(wcss)
```

```python
plt.plot(range(1,11),wcss)
plt.show()
```

```python
model = KMeans(n_clusters=4)
prediction=model.fit_predict(X)
print(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=prediction)
plt.show()
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title("dendrogram")
dend = shc.dendrogram(shc.linkage(X,method = 'ward'))
plt.show()
```

```python
Ag = AgglomerativeClustering(n_clusters = 4, metric='euclidean', linkage='ward')
prediction1 = Ag.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c = prediction1)
plt.show()
```
